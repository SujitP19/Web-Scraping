```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
URL = 'https://www.amazon.in/product-reviews/8172234988/ref=acr_dp_hist_5?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews#reviews'
```

```python
page = requests.get(URL)
```

```python
soup = bs(page.content, 'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('span', class_ = "a-profile-name")
```

```python
print(names)
```

```python
cust_names = []

for name in range(len(names)):
    
    cust_names.append(names[name].get_text())
```

```python
len(cust_names)
```

```python
len(set(cust_names))
```

```python
cust_names
```

```python
title = soup.find_all('a', class_ ='review-title-content')
```

```python
len(cust_names)
```

```python
title
```

```python
cust_titles = []

for name in range(len(title)):
    
    cust_titles.append(title[name].get_text())
```

```python
cust_titles
```

```python
cust_titles[:] = [titles.lstrip('\n') for titles in cust_titles]
```

```python
len(cust_titles)
```

```python
cust_titles[:] = [titles.rstrip('\n') for titles in cust_titles]
```

```python
cust_titles
```

```python
star_ratings = soup.find_all('i', class_ = 'review-rating')
```

```python
star_ratings
```

```python
ratings = []

for name in range(len(star_ratings)):
    
    ratings.append(star_ratings[name].get_text())
```

```python
len(ratings)
```

```python
ratings
```

```python
review = soup.find_all('span', {'data-hook':'review-body'})
```

```python
review
```

```python
reviews = []

for name in range(len(review)):
    
    reviews.append(review[name].get_text())
```

```python
reviews
```

```python
reviews[:] = [review.rstrip('\n') for review in reviews]
```

```python
reviews
```

```python
reviews[:] = [review.lstrip('\n') for review in reviews]
```

```python
reviews
```

```python
len(reviews)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['CUSTOMERS'] = cust_names
df['REVIEW TITLE'] = cust_titles
df['RATINGS'] = ratings 
df['REVIEWS'] = reviews
```
